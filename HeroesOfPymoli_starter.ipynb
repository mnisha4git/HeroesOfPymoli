## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
item_count = purchase_data["Item ID"].nunique()
average_price = round(purchase_data["Price"].mean(),2)
num_purchases = purchase_data["Purchase ID"].nunique() 
total_revenue = round(purchase_data["Price"].sum(),2)

purchase_details = [{"Average Price": average_price ,"Number of Purchases": num_purchases,"Number of Unique Items": item_count,"Total Revenue":total_revenue}]
purchase_summary = pd.DataFrame(purchase_details)
organized_df = purchase_summary[["Average Price","Number of Purchases","Number of Unique Items","Total Revenue"]]
organized_df
```
